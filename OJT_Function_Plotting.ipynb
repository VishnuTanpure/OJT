## Scatter Plot

````python
# X-coordinates
X = np.random.randint(-4,5,10)
X
````

````python
# Y-coordinates
Y = np.random.randint(-4,5,10)
Y
````

````python
#Scatter Plot
plt.scatter(X,Y)
````

````python
# Change colour of points
plt.scatter(X,Y, color='red')

# labeling the axes
plt.xlabel('age')
plt.ylabel('salary')

# Show the plot
plt.show()
````

````python
#Line Plot
plt.scatter(X,Y, color='red')

# plt.plot() will plot a line
plt.plot(X,Y, color='blue')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

````python
X = np.array([-1,-1,1,1,-1])
Y = np.array([-1,1,1,-1,-1])
plt.plot(X,Y, linestyle='-.')
````

## Plot a function $y = f(x)$

Plot: $y=x^2$

````python
# Create x-coordinates (Domain)
X = np.arange(-2, 2.1, 0.1)
X
````

````python
3**2
````

````python
# plot y=x^2
plt.plot(X,X**2, color='blue')

plt.xlabel('x')
plt.ylabel(r'$x^2$')
plt.show()
````

### Add text in figure

````python
plt.plot(X,X**2, color='blue')

plt.text(-0.5, 3, 'Parabola $Y = x^2$', fontsize = 10) # Add text in figure

plt.xlabel('x')
plt.ylabel(r'$x^2$')
plt.show()
````

### Annotate figure

````python
plt.plot(X,X**2, color='blue')
plt.text(0, 2, 'Parabola $Y = x^2$', fontsize = 10)

#Annotate
plt.annotate('loc', xy=(2, 2), xytext=(0, 1), arrowprops=dict(facecolor='green', shrink=0.0))


plt.xlabel('x')
plt.ylabel(r'$x^2$')
plt.show()
````

###Combine two Plots

````python
X = np.arange(-2*np.pi, 2*np.pi+0.1, 0.1)
````

````python
plt.plot(X,np.sin(X), color='blue')
plt.plot(X,np.cos(X), color='red')
plt.plot(np.array([-8,8]),np.array([0,0]), color='black') #X-axis
plt.plot(np.array([0,0]),np.array([-1,1]), color='black') #Y-axis


plt.xlabel('x')
plt.ylabel(r'$x^2$')
plt.show()
````

### Add legend

````python
plt.plot(X,np.sin(X), color='blue')
plt.plot(X,np.cos(X), color='red')

#Legend
plt.legend(['sin(x)','cos(x)'], loc="lower left")

plt.xlabel('x')
plt.ylabel(r'$y$')

plt.show()
````

Add Grid in the figure

````python
plt.plot(X,np.sin(X), color='blue')
plt.plot(X,np.cos(X), color='red')

# Legend
plt.legend(['sin(x)','cos(x)'], loc="upper right")

plt.xlabel('x')
plt.ylabel(r'$x^2$')

# Grid
plt.grid()

plt.show()
````

###Display Multiple Plots
With the subplot() function you can draw multiple plots in one figure:



```
plt.subplot(1, 2, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
```



````python
plt.figure()
plt.subplot(2,1,1)
plt.plot(X, np.exp(X), color='red')

plt.subplot(2,1,2)
plt.plot(X, np.sin(X), color='blue')

plt.show()
````

````python
plt.figure()
plt.subplot(1,3,1)
plt.plot(X, np.exp(X), color='red')

plt.subplot(1,3,2)
plt.plot(X, np.sin(X), color='blue')

plt.subplot(1,3,3)
plt.plot(X, np.cos(X), color='g')

plt.show()
````

### Change figure size

````python
plt.figure(figsize=(15, 8), dpi=180, facecolor='w', edgecolor='r')

plt.subplot(1,3,1)
plt.plot(X, np.exp(X), color='red', label = 'A')

plt.subplot(1,3,2)
plt.plot(X, np.sin(X), color='blue')

plt.subplot(1,3,3)
plt.plot(X, np.cos(X), color='g')

plt.show()
````
